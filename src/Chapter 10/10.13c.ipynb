```python
from vpython import graph,gdots,color,vector
from numpy import zeros,pi,cos,sin,sqrt
from random import randrange,random

#função que realiza a mudança de posiçao da particula
def passo(x,y):
    p = randrange(4)
    if p==0: #passo para cima
        if y != (L-1):
            return x, y+1
        else:
            return passo(x,y)
    elif p==1: #passo para direita
        if x != (L-1):
            return x+1, y
        else:
            return passo(x,y)
    elif p==2: #passo para baixo
        if y != 0:
            return x, y-1
        else:
            return passo(x,y)
    elif p==3: #passo para esquerda
        if x != 0:
            return x-1, y
        else:
            return passo(x,y) 

#parametros
L = 101 #tamanho do reticulado; obs:deve ser um valor impar
x0 = int((L-1)/2) #posicao x
y0 = int((L-1)/2) #posicao y

#matriz que salva quais posições estao ocupadas:0 desocupada, 1 ocupada
pos = zeros([L,L], int)

#cria o background da animação e a lista em que as parituclas serão salvas
g = graph(width=600, height=600, foreground=color.black, background=color.white, xmin=0, xmax=L-1, ymin=0, ymax=L-1)
particulas = []

#cria uma primeira particula no centro
particulas.append(gdots(pos=(x0,y0), radius = 3))
pos[x0][y0] = 1

r = 0
k = 0
while r<(L-1)/4:
    #escolhe uma posição angular aleatoria para a particula iniciar
    theta = 2*pi*random()
    x = (r+1)*cos(theta) + x0
    y = (r+1)*sin(theta) + y0
    
    #vamos aproximar os valores para o inteiro maior mais proximo
    x = int((x // 1) + 1)
    y = int((y // 1) + 1)
    
    anchored = 0
    while anchored==0:
        if (r!=0 and 2*r < sqrt((x-x0)**2 + (y-y0)**2)) or (r==0 and sqrt((x-x0)**2 + (y-y0)**2)>2):#particula passou dos limiste
            break
        
        if x==0:
            if (pos[x+1][y]==1 or pos[x][y+1]==1 or pos[x][y-1]==1):
                pos[x][y] = 1
                anchored = 1
        elif x==L-1:
            if (pos[x-1][y]==1 or pos[x][y+1]==1 or pos[x][y-1]==1):
                pos[x][y] = 1
                anchored = 1
        elif y==0:
            if (pos[x-1][y]==1 or pos[x][y+1]==1 or pos[x+1][y]==1):
                pos[x][y] = 1
                anchored = 1    
        elif y==L-1:
            if (pos[x-1][y]==1 or pos[x+1][y]==1 or pos[x][y-1]==1):
                pos[x][y] = 1
                anchored = 1              
        elif pos[x+1][y]==1 or pos[x-1][y]==1 or pos[x][y+1]==1 or pos[x][y-1]==1: #verifica ha particulas na vizinhança
            pos[x][y] = 1
            anchored = 1
        
        if anchored == 0: 
            x, y = passo(x,y)
        
    
    if anchored == 1:
        particulas.append(gdots(pos=(x,y), radius = 3, color=vector(0,0,k/350)))
        if sqrt((x-x0)**2 + (y-y0)**2) > r:
            r = sqrt((x-x0)**2 + (y-y0)**2)
        k += 1
    
print(k)
```
